To find the total distance between the left list and the right list, add up the distances between all of the pairs you found. In the example above, this is 2 + 1 + 0 + 1 + 2 + 5, a total distance of 11!

```python
import re

f = open('Problem 1.txt','r')

list1 = []
list2 = []

for line in f:
    nums = re.findall('[0-9]+',line)
    list1.append(nums[0])
    list2.append(nums[1])

print(list1)
print(list2)
```

The list has now been imported and has been split up into two lists. It is now time to quickly program up some bubble sort.

```python
def bubblesort(list):
    swapped = True
    while swapped == True:
        swapped = False
        for i in range(len(list)-1):
            if list[i]>list[i+1]:
                x = list[i]
                list[i] = list[i+1]
                list[i+1] = x
                swapped = True
    return list
```

Now lets sort some numbers.

```python
list1 = bubblesort(list1)
print(list1)
list2 = bubblesort(list2)
print(list2)
```

```python
import numpy as np

total = 0
for i in range(len(list2)):
    total += np.abs(int(list1[i])-int(list2[i]))

print(total)
```

```python
similarity = 0
for i in range(len(list1)-1):
    ctr = 0
    enum = 0

    while int(list2[enum]) <= int(list1[i]) and enum < len(list2)-1:
        if list2[enum] == list1[i]:
            ctr += 1
        enum += 1
    similarity += int(list1[i])*ctr

print(similarity)
```
